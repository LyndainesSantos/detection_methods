```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Carregar a imagem
img = cv2.imread('/content/images/q01.PNG', cv2.IMREAD_GRAYSCALE)
```

```python
def img_binary(image):
    _, image_binary = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
    return image_binary
def preprocessing_binary(imagem):
    # Garantir que a imagem é binária (0s e 255s)
    image_binary = img_binary(img)

    contours_binary, _ = cv2.findContours(image_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    print("Quantidade de objetos na imagem: ", len(contours_binary))

    return contours_binary

def preprocessing_gauss(image):

    # Aplicar um filtro Gaussiano para suavizar a imagem e reduzir o ruído
    image_gauss = cv2.GaussianBlur(image, (5, 5), 0)

    image_binary = img_binary(image_gauss)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
    image_morphology = cv2.morphologyEx(image_binary, cv2.MORPH_OPEN, kernel)

    contours_gauss, _ = cv2.findContours(image_morphology, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    print("Quantidade de objetos na imagem: ", len(contours_gauss))

    return contours_gauss

def draw_contours(image, contours):
    # Desenhar os contornos na imagem
    image_contours = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)  # Convertendo para RGB para desenhar em cores
    cv2.drawContours(image_contours, contours, -1, (0, 255, 0), 2)
    return image_contours

def display_img(image, position, title):
    # Mostrar a imagem
    plt.subplot(1, 3, position)
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
```

```python
contoursBinary = preprocessing_binary(img)
contoursGauss = preprocessing_gauss(img)
img_0 = draw_contours(img, contoursBinary)
img_1 = draw_contours(img, contoursGauss)
```

```python
plt.figure(figsize=(10, 8))
display_img(img, 1, 'Imagem Original')
display_img(img_0, 2, f'Imagem sem filtro \n{len(contoursBinary)} objetos detectados')
display_img(img_1, 3, f'Imagem com filtro \n{len(contoursGauss)} objetos detectados')
plt.show()
```
